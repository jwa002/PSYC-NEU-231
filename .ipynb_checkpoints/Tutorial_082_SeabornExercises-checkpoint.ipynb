# Seaborn Exercises

1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#add a percentage tips column
tips = pd.DataFrame(tips)
tips['Percentage Tip']=((tips['tip']/tips['total_bill'])*100)
tips.head()
```

```python
#visualize relationship between tip percentage, day of the week, and time
sns.boxplot(x='day', y='Percentage Tip', data=tips, hue='time')
plt.title('Tip Percentage by Day and Time')
plt.show()
```

```python
#visualize tip by day, time, and number of people
g = sns.PairGrid(tips, y_vars="Percentage Tip", x_vars =["day","time","size"], height = 3)
g.map(sns.pointplot)
plt.show()
```

2) Find an interesting trend in the "exercise" dataset

```python
#import "exercise" dataset
exercise = sns.load_dataset('exercise')
exercise.head()
```

```python
#find interesting trend
sns.pointplot(x='time',y='pulse',data=exercise, hue='kind')
plt.title('Pulse by Activity Type and Duration')
plt.show()

sns.pointplot(x='kind',y='pulse',data=exercise,hue='diet')
plt.show()

sns.catplot(x='time', y='pulse', hue='kind', col='diet', data=exercise, kind='point')
plt.show()
```
